```r
library(grid)
library(vcd)
library(ca)
```

```r
data(Arthritis)
help(Arthritis)
```

```r

Arthritis.use <- Arthritis[,c("Treatment", "Sex", "Age", "Improved")]
Arthritis.use[,"Age"] <- cut(Arthritis.use[,"Age"], seq(min(Arthritis.use[,"Age"]), 
                                                   max(Arthritis.use[,"Age"]), 
                                                   length.out = 7))
head(Arthritis.use)
```

```r
(Arthritis.use.table <- with(Arthritis.use, table(Treatment, Improved)))
```

```r
Arthritis.use.ca <- ca(Arthritis.use.table)
Arthritis.use.ca$rowcoord
Arthritis.use.ca$colcoord
```

```r
plot(c(Arthritis.use.ca$rowcoord, Arthritis.use.ca$colcoord), 
     rep(0,5), col = c(rep(2,2), rep(3,3)),
     pch =20)
text(Arthritis.use.ca$colcoord,  rep(0,3), 
     rownames(Arthritis.use.ca$colcoord),
      cex= 0.7, pos=3)
text(Arthritis.use.ca$rowcoord,  rep(0,2), 
     rownames(Arthritis.use.ca$rowcoord),
      cex= 0.7, pos=1)
```
